# Automate files aggregation

```python
import pandas as pd
file = 'gwz_orders_samples.csv'
df_orders = pd.read_csv(file)
df_orders.head()
```

```python
import os

# current working directory
root = os.getcwd()

# sales folder path
sales_folder_path = os.path.join(root, "sales")

# create sales folder
os.mkdir(sales_folder_path)

# list all unique dates in orders
for date in df_orders["date_date"].unique():

  # create date folder
  sales_date_path = os.path.join(root, "sales", str(date))
  os.mkdir(sales_date_path)

  # save orders for that date
  df_sales_date = df_orders[df_orders.date_date==str(date)].reset_index()
  df_sales_date.to_csv(os.path.join(sales_date_path, "orders.csv"), index=False)
```

```python
all_orders = []

for date_folder in os.listdir(sales_folder_path):

  # build path as all files are sales_folder_path, date_folder, "orders.csv"
  file_path = os.path.join(sales_folder_path,date_folder,'orders.csv')

  # load data in DataFrame
  df_date_folder = pd.read_csv(file_path)

  # append to the list
  all_orders.append(df_date_folder)

  df_orders = pd.concat(all_orders, axis=0)
  df_orders.shape
```
